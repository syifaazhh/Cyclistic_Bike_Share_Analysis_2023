# Cyclistic Bike Share Analysis 2023

# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import datetime
import os

print("Bravo, libraries imported successfully!")
```

# 2. Import Datasets

```python
jan_23 = pd.read_csv("202301-divvy-tripdata.csv")
feb_23 = pd.read_csv("202302-divvy-tripdata.csv")
mar_23 = pd.read_csv("202303-divvy-tripdata.csv")
apr_23 = pd.read_csv("202304-divvy-tripdata.csv")
may_23 = pd.read_csv("202305-divvy-tripdata.csv")
jun_23 = pd.read_csv("202306-divvy-tripdata.csv")
jul_23 = pd.read_csv("202307-divvy-tripdata.csv")
aug_23 = pd.read_csv("202308-divvy-tripdata.csv")
sep_23 = pd.read_csv("202309-divvy-tripdata.csv")
oct_23 = pd.read_csv("202310-divvy-tripdata.csv")
nov_23 = pd.read_csv("202311-divvy-tripdata.csv")
dec_23 = pd.read_csv("202312-divvy-tripdata.csv")

print("Yay, CSV files are imported successfully!")
```

# 3. Merge The Datasets

```python
dataset = [jan_23, feb_23, mar_23, apr_23, may_23, jun_23, jul_23, aug_23, sep_23, oct_23, nov_23, dec_23]
dataset = pd.concat(dataset)

print("Datasets are merged successfully!")

dataset
```

# 4. Checking The Data Types

```python
dataset.info()
```

# 5. Checking The Null Values

```python
dataset.isnull().sum()
```

# 6. Remove The Null Values

```python
dataset.dropna(how = 'any', axis = 0, inplace = True)

print("Null values removed successfully!")
print(dataset.isnull().sum())
```

# 7. Checking The Duplicate Values

```python
if dataset.duplicated().any():
    print("There are duplicate in the datasets!")
else:
    print("There are no duplicate in the datasets!")
```

# 8. Change The Data Type to Date and Time

```python
pd.options.mode.chained_assignment = None

dataset['started_at'] = pd.to_datetime(dataset['started_at'])
dataset['ended_at'] = pd.to_datetime(dataset['ended_at'])
print("Started_at and Ended_at variables are changed successfully!")

dataset.info()
```

# 9. Making a New Column for Hours

```python
dataset["start_hour"] = dataset["started_at"].dt.hour
dataset["start_hour"].unique()
```

# 10. Making a New Column for Days

```python
dataset["start_day_num"] = dataset["started_at"].dt.dayofweek
day_dict = {0: "Monday", 1: "Tuesday", 2:"Wednesday", 3: "Thursday", 4: "Friday", 5: "Saturday", 6: "Sunday"}
dataset["start_day_name"] = dataset["start_day_num"].apply(lambda y:day_dict[y])
dataset["start_day_name"].unique()
```

```python
dataset["day_of_year"] = dataset["started_at"].dt.day_of_year
```

# 11. Making a New Column for Months

```python
dataset["start_month_num"] = dataset["started_at"].dt.month
dataset["start_month_num"].unique()
```

```python
month_dict = {1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 12:"December"}
dataset["start_month_name"]= dataset["start_month_num"].apply(lambda y:month_dict[y])
dataset["start_month_name"].unique()
```

# 12. Making a New Column for Years

```python
dataset["start_year"] = dataset["started_at"].dt.year
dataset["start_year"].unique()
```

# 13. Making a New Column for Ride Length

```python
dataset["ride_length"] = dataset["ended_at"] - dataset["started_at"]
dataset["ride_length"] = pd.to_numeric(dataset["ride_length"])/6e+10
```

```python
dataset["ride_distance"] = np.sqrt(((dataset["end_lat"] - dataset["start_lat"])**2) + ((dataset["end_lng"] - dataset["start_lng"])**2))
dataset["ride_distance"] = dataset["ride_distance"] * 111.139
```

# 14. Rename Column for Member Casual

```python
dataset.rename(columns = {'member_casual': 'user_type'}, inplace=True)
dataset.info()
```

# 15. Making Values for Weekend and Weekdays

```python
dataset["weekend"] = np.where(dataset["start_day_num"] > 4, True, False)
dataset.info()
```

```python
type_of_week = dataset[['start_day_name', 'weekend']]
type_of_week
```

```python
dataset.sort_values(by = ["started_at"], inplace = True, ascending = True)
dataset
```

# 16. Describing Ride Length

```python
dataset["ride_length"].describe(include = "all")
```

```python
trip_mins = dataset[["started_at", "ended_at", "ride_length"]]
trip_mins.nlargest(10, ["ride_length"])
```

```python
#Remove the outliers
outliers_ride_length = (dataset["ride_length"] < 1) | (dataset["ride_length"] > 480)
outliers_ride_length.value_counts()
```

```python
dataset.drop(dataset.index[outliers_ride_length], inplace = True)

print("Inconsistent values are deleted successfully!")
```
